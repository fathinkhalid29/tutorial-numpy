```python
matrix_arr  = np.array([[3,4,5],[6,7,8],[9,5,45]])
print("Original matrix array: {}".format(matrix_arr))
```

```python
print("slices the first two rows: {}".format(matrix_arr[:2]))
```

```python
print("slices the first two rows and two columns: {}".format(matrix_arr[:2, 1:]))
```

```python
print("return 3 and 4: {}".format(matrix_arr[0, 1:]))
```

```python
print("return 3 and 4: {}".format(matrix_arr[0, :2]))
```
